## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Research Question

Which content pages should be reviewed first for refresh, expansion, protection, pruning, or monitoring based on historical search and engagement signals?

Decision

Which pages should the SEO or content team prioritize for review?

Who acts?

SEO managers or content editors.

Action

Review the highest-ranked pages and decide whether to refresh, expand, protect, prune, or monitor them.

Cost of a Wrong Recommendation
False Positive: A healthy page is recommended for review, wasting editor time and review capacity.
False Negative: A page that truly needs attention is missed, allowing traffic or engagement to continue declining.

Why Machine Learning?

Page performance depends on many interacting signals such as impressions, clicks, CTR, average position, engagement rate, freshness, and content characteristics. These relationships are difficult to capture using a single rule, so machine learning can help prioritize pages for human review.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# -------------------------------
# Basic Information
# -------------------------------
print("=" * 60)
print("DATASET OVERVIEW")
print("=" * 60)
print(f"Rows: {len(df):,}")
print(f"Columns: {len(df.columns)}")
print(f"Unique clients: {df['client_id'].nunique()}")

# -------------------------------
# Content Type Distribution
# -------------------------------
print("\n" + "=" * 60)
print("CONTENT TYPES")
print("=" * 60)
print(df["content_type"].value_counts())

# -------------------------------
# Trend Direction Distribution
# -------------------------------
print("\n" + "=" * 60)
print("TREND DIRECTION")
print("=" * 60)
print(df["trend_direction"].value_counts())

print("\nTrend Direction (%)")
print((df["trend_direction"].value_counts(normalize=True) * 100).round(2))

# -------------------------------
# Performance Statistics
# -------------------------------
print("\n" + "=" * 60)
print("PERFORMANCE STATISTICS")
print("=" * 60)

print(f"Average 90-day Impressions : {df['impressions_90d'].mean():.2f}")
print(f"Average 90-day Clicks      : {df['clicks_90d'].mean():.2f}")
print(f"Average CTR (%)            : {df['ctr'].mean():.2f}")

# Average Position (0 = no data)
avg_position = df["avg_position"].replace(0, pd.NA).mean()
print(f"Average Search Position    : {avg_position:.2f}")

print(f"Average Engagement Rate (%) : {df['engagement_rate'].mean():.2f}")
print(f"Average AI Traffic (%)      : {df['ai_traffic_pct'].mean():.2f}")

# -------------------------------
# Missing Values
# -------------------------------
print("\n" + "=" * 60)
print("MISSING VALUES")
print("=" * 60)
print(df.isnull().sum().sort_values(ascending=False).head(10))
```

The starter dataset contains 30,000 content pages described by 44 features across 32 pseudonymized clients. Most pages (27,207, about 90.7%) are keyword articles, while 2,096 are Feedly articles and 697 are comparison articles. The dataset also shows substantial variation in page performance: 16,262 pages are currently marked as "down" in trend_direction, indicating many pages may benefit from review. These observations suggest that ranking pages by their likelihood of needing attention is a meaningful problem for content teams, making the Refresh / Content Opportunity Scoring lane worth investigating over the next seven weeks.
